```python
import requests
import os
from bs4 import BeautifulSoup
```

```python
url = 'https://bdshanoi.com.vn/can-ho/next-page-2.html'
```

```python
so_luong_dong = 1000
```

```python
data = []
```

```python
while len(data) < so_luong_dong:

    response = requests.get(url)
    if response.status_code == 200:

        soup = BeautifulSoup(response.content.decode('utf-8', 'ignore'), 'html.parser')

        lines = soup.find_all(['h4', 'p', 'span'])

        for line in lines:
            data.append(line.text.strip())
    else:
        print("KhÃ´ng láº¥y Ä‘Æ°á»£c dá»¯ liá»‡u")
        break
```

```python
data = data[:so_luong_dong]
```

```python
for i, line in enumerate(data):
    print(f"{i+1}. {line}")
```

```python

with open('data.txt', 'w', encoding='utf-8') as file:
    for line in data:
        file.write(line + '\n')
```
